```python
# Reading the new file in which all files are merged

all_data=pd.read_csv("total_data.csv")
all_data
```

```python
# This line will tell you how much rows were having null values
nan= all_data[all_data.isna().any(axis=1)]
nan
```

```python
# Removing the Null Values
all_data = all_data.dropna(how='all')
all_data
```

```python
# To find if there is any-any null value in your dataset
all_data.isnull().any().any()
```

```python
# Finding 'Or' letter in order date cause when converting month intoint data type it shows error invalid literal 'Or'
temp_df=all_data[all_data['Order Date'].str[0:2]=='Or']
temp_df
```

```python
# Here getting only those columnn which don't have 'Or' in columns
all_data=all_data[all_data['Order Date'].str[0:2]!='Or']
all_data
```

```python
# Since we find that there is no specific/separate month column for calculations. So,
#Adding Month Column for monthly sales and converting it into int datatype

all_data['Month']= all_data['Order Date'].str[0:2]
all_data['Month']=all_data['Month'].astype('int32')
all_data


```

#### Question 1: What was the best month for sales? How much was earned that month? 

```python
# First Converting str values in quantity covered and price each into numeric
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
```

```python
#### Adding Sales column
all_data['Sales']=all_data['Quantity Ordered']* all_data['Price Each']
all_data.head()
```

```python
# Answering the above Question
results = all_data.groupby('Month').sum()
results.head()
```

```python
import matplotlib.pyplot as plt

months=range(1,13)
# plt.figure(figsize=(9, 7))

bars=plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sum of Sales in month')
plt.xlabel('Months')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', rotation=90)

plt.show()
```

From the above graph  
I conclude the month december has more sales than any other month. this could be any reason why this december has more sales...

#### Question 4: What products are most often sold together?


```python
all_data.head()
```

```python
#getting duplicate id so that we know nwhch product are sold together
df=all_data[all_data['Order ID'].duplicated(keep = False)]
df['Grouped Product']=df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df=df[['Order ID','Grouped Product']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count= Counter()

for row in df['Grouped Product']:
    row_list= row.split(',')
    count.update(Counter(combinations(row_list,2)))


for key, value in count.most_common(10):
    print(key, value)
```

From the above calculation i found that the most sold item is Iphone and Lightning charging cable and the sold amount of these twop together is 1005 
So to advertise the cable with the iphone will be best suited to company to sold more product of this...
